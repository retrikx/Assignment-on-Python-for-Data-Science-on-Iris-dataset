```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")




```

```python
df.groupby('Classification').size()
```

```python
sns.FacetGrid(df, hue="Classification", palette="husl",size=5) \
   .map(plt.scatter, "SL", "SW") \
   .add_legend()
```

```python
sns.stripplot(x='PL',y='SL',hue='Classification',data=df)
sns.set(rc={"figure.figsize":(15,6)})
plt.title('Comparison of SL and PL of each flower', fontsize=20)
plt.show()
```

```python
df[['SL','SW','PL','PW']].plot.hist()
```
